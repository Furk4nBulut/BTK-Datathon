```python
import re
# Tarihlerin normal olduğu durumu işlemeye çalış
def extract_year(date_string):
    try:
        # Tarih formatında yıl çıkarma (tarih doğru formatta olursa)
        return pd.to_datetime(date_string).year
    except (ValueError, TypeError):
        # Eğer format uygun değilse, son dört basamaklı sayıyı yıl olarak al
        match = re.search(r'\b(19|20)\d{2}\b', date_string)
        if match:
            return int(match.group(0))
        return None  # Yıl bulunamazsa None döndür

# 'Dogum Tarihi' sütunundan yıl çıkartma
data['Dogum Tarihi'] = data['Dogum Tarihi'].apply(extract_year)
# Sonuçları görüntüleme
print(data[['Dogum Tarihi', 'Dogum Tarihi']])
```

```python
def update_city_info(city_list):
    # Türkiye'nin 81 ilinin listesi
    cities_in_turkey = [
        "Adana", "Adıyaman", "Afyonkarahisar", "Ağrı", "Aksaray", "Amasya", "Ankara", "Antalya", "Ardahan", 
        "Artvin", "Aydın", "Balıkesir", "Bartın", "Batman", "Bayburt", "Bilecik", "Bingöl", "Bitlis", "Bolu", 
        "Burdur", "Bursa", "Çanakkale", "Çankırı", "Çorum", "Denizli", "Diyarbakır", "Düzce", "Edirne", 
        "Elazığ", "Erzincan", "Erzurum", "Eskişehir", "Gaziantep", "Giresun", "Gümüşhane", "Hakkari", "Hatay", 
        "Iğdır", "Isparta", "İstanbul", "İzmir", "Kahramanmaraş", "Karabük", "Karaman", "Kars", "Kastamonu", 
        "Kayseri", "Kırıkkale", "Kırklareli", "Kırşehir", "Kilis", "Kocaeli", "Konya", "Kütahya", "Malatya", 
        "Manisa", "Mardin", "Mersin", "Muğla", "Muş", "Nevşehir", "Niğde", "Ordu", "Osmaniye", "Rize", "Sakarya", 
        "Samsun", "Siirt", "Sinop", "Sivas", "Şanlıurfa", "Şırnak", "Tekirdağ", "Tokat", "Trabzon", "Tunceli", 
        "Uşak", "Van", "Yalova", "Yozgat", "Zonguldak"
    ]
    
    updated_list = []
    
    for item in city_list:
        # Veriyi şehir ve bölge olarak ayırıyoruz, eğer ikiye bölünmezse default olarak region'u boş bırakıyoruz
        if ',' in item:
            place, region = item.split(',', 1)  # İlk virgülden ayırıyoruz, geri kalanı 'region' olarak alıyoruz
        else:
            place, region = item, ""  # Eğer virgül yoksa sadece 'place' var, 'region' boş kalır

        # Şehir ismi Türkiye'nin 81 ili arasında mı kontrol ediyoruz
        updated_place = None
        for city in cities_in_turkey:
            if city.lower() in place.lower():
                updated_place = city
                break

        # Eğer bulunduysa güncellenmiş şehir ismini ekliyoruz
        if updated_place:
            updated_list.append(f"{updated_place}")
        else:
            updated_list.append(item)  # Bulunmazsa aynı bırakıyoruz
    
    return updated_list

# Şehir isimlerini güncelleme işlemi
data['Ikametgah Sehri'] = update_city_info(data['Ikametgah Sehri'].astype(str))
data['Dogum Yeri'] = update_city_info(data['Dogum Yeri'].astype(str))
data['Lise Sehir'] = update_city_info(data['Lise Sehir'].astype(str))

# Sonuçları görüntüleyelim
data[['Ikametgah Sehri', 'Dogum Yeri', 'Lise Sehir']]

```

```python
# Türkçe karakterleri İngilizce karşılıklarıyla değiştiren bir sözlük tanımlayın
turkish_to_english = str.maketrans(
    'çğıöşüÇĞİÖŞÜ',
    'cgiosuCGIOSU'
)

# 'Bölüm' sütunundaki Türkçe karakterleri İngilizce karşılıklarıyla değiştirin
data['Bölüm'] = data['Bölüm'].str.translate(turkish_to_english)

# Veriyi küçük harfe dönüştürün
data['Bölüm'] = data['Bölüm'].str.lower()


#unique values
data['Bölüm'].value_counts()
```
