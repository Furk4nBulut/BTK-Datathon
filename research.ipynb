```python
# import pandas
import pandas as pd
```

```python
# Process the Gender column: Convert 'KADIN' to 1, otherwise 0
cleaned_train_data['Cinsiyet'] = cleaned_train_data['Cinsiyet'].apply(
    lambda x: 1 if isinstance(x, str) and x.strip().upper() == 'KADIN' else 0
)

# Process the Scholarship column: Convert 'HAYIR' to 1, otherwise 0
cleaned_train_data['Burs Aliyor mu?'] = cleaned_train_data['Burs Aliyor mu?'].apply(
    lambda x: 1 if isinstance(x, str) and x.strip().upper() == 'EVET' else 0
)

# Process the University Type column: Convert 'DEVLET' to 1, otherwise 0
cleaned_train_data['Universite Turu'] = cleaned_train_data['Universite Turu'].apply(
    lambda x: 1 if isinstance(x, str) and x.strip().upper() == 'DEVLET' else 0
)

# Process the Entrepreneurship Club membership column: Convert 'EVET' to 1, otherwise 0
cleaned_train_data['Girisimcilik Kulupleri Tarzi Bir Kulube Uye misiniz?'] = cleaned_train_data['Girisimcilik Kulupleri Tarzi Bir Kulube Uye misiniz?'].apply(
    lambda x: 1 if isinstance(x, str) and x.strip().upper() == 'EVET' else 0
)

# Process the Previous University Graduation column: Convert 'EVET' to 1, otherwise 0
cleaned_train_data['Daha Once Baska Bir Universiteden Mezun Olmus'] = cleaned_train_data['Daha Once Baska Bir Universiteden Mezun Olmus'].apply(
    lambda x: 1 if isinstance(x, str) and x.strip().upper() == 'EVET' else 0
)

# Process the Previous University Graduation column: Convert 'EVET' to 1, otherwise 0
cleaned_train_data['Daha Once Baska Bir Universiteden Mezun Olmus'] = cleaned_train_data['Daha Once Baska Bir Universiteden Mezun Olmus'].apply(
    lambda x: 1 if isinstance(x, str) and x.strip().upper() == 'EVET' else 0
)

# Process the Professional Sports Engagement column: Convert 'EVET' to 1, otherwise 0
cleaned_train_data['Profesyonel Bir Spor Daliyla Mesgul musunuz?'] = cleaned_train_data['Profesyonel Bir Spor Daliyla Mesgul musunuz?'].apply(
    lambda x: 1 if isinstance(x, str) and x.strip().upper() == 'EVET' else 0
)

# Process the Active NGO Membership column: Convert 'EVET' to 1, otherwise 0
cleaned_train_data['Aktif olarak bir STK üyesi misiniz?'] = cleaned_train_data['Aktif olarak bir STK üyesi misiniz?'].apply(
    lambda x: 1 if isinstance(x, str) and x.strip().upper() == 'EVET' else 0
)

# Process the Active NGO Membership column: Convert 'EVET' to 1, otherwise 0
cleaned_train_data['Girisimcilikle Ilgili Deneyiminiz Var Mi?'] = cleaned_train_data['Girisimcilikle Ilgili Deneyiminiz Var Mi?'].apply(
    lambda x: 1 if isinstance(x, str) and x.strip().upper() == 'EVET' else 0
)

# Define the encoding function with space handling
def encode_gpa(gpa):
    gpa = str(gpa).replace(" ", "").strip()  # Remove spaces and strip any extra whitespace
    if gpa == '3.00-4.00':
        return 4
    elif gpa == '3.50-3':
        return 3
    elif gpa == '3.00-2.50':
        return 2
    elif gpa == '2.50vealtı':  # Boşluklar kaldırıldığı için bu şekilde yazıldı
        return 1
    else:
        return None  # For unexpected values

# Apply the encoding function directly to the existing 'Universite Not Ortalamasi' column
cleaned_train_data['Universite Not Ortalamasi'] = cleaned_train_data['Universite Not Ortalamasi'].apply(encode_gpa)

# To verify the results, display the full dataframe or the first few rows
cleaned_train_data.head()
```

```python
# First, normalize all values (convert to lowercase and remove spaces)
def clean_education(value):
    if pd.isna(value) or value == '0':  # If the value is NaN or '0', return None
        return None
    elif isinstance(value, str):  # If the value is a string
        value = value.lower().strip().replace(' ', '')
        # Normalize different variations of the same category
        if value in ['üniversite', 'üniversitemezunu', 'universite']:
            return 'universite'
        elif value in ['yükseklisans', 'yükseklisans/doktora', 'yükseklisans/doktara']:
            return 'yukseklisans'
        elif value in ['ilkokul', 'ilkokulmezunu', 'i̇lkokulmezunu']:
            return 'ilkokul'
        elif value in ['ortaokul', 'ortaokulmezunu', 'ortaokul']:
            return 'ortaokul'
        elif value in ['lise', 'lisemezunu', 'lise']:
            return 'lise'
        elif value in ['doktora', 'doktara']:
            return 'doktora'
        elif value in ['eğitimyok', 'egitimiyok', 'eğitimyok']:
            return 'egitimyok'
    return value  # Return the value as it is if it's already clean

# Define the encoding based on the cleaned categories
education_mapping = {
    'ilkokul': 1,
    'ortaokul': 2,
    'lise': 3,
    'universite': 4,
    'yukseklisans': 5,
    'doktora': 6,
    'egitimyok': 0
}

# Apply the cleaning function first, and then map the cleaned values to numeric
cleaned_train_data['Anne Egitim Durumu'] = cleaned_train_data['Anne Egitim Durumu'].apply(clean_education).map(education_mapping)
cleaned_train_data['Baba Egitim Durumu'] = cleaned_train_data['Baba Egitim Durumu'].apply(clean_education).map(education_mapping)
```

```python
# Define the mapping function for 'Baba Calisma Durumu'
def encode_work_status(value):
    if pd.isna(value):  # If the value is NaN, return None
        return None
    value = value.strip().lower()  # Normalize the string (lowercase, remove spaces)
    if value == 'evet':
        return 1
    elif value == 'hayır':
        return 0
    elif value == 'emekli':
        return 2
    else:
        return None  # Return None for unexpected values

# Apply the encoding to the 'Anne Calisma Durumu' and'Baba Calisma Durumu' columns
cleaned_train_data['Anne Calisma Durumu'] = cleaned_train_data['Anne Calisma Durumu'].apply(encode_work_status)
cleaned_train_data['Baba Calisma Durumu'] = cleaned_train_data['Baba Calisma Durumu'].apply(encode_work_status)
```

```python
# Define a function to encode 'Lise Mezuniyet Notu' combining GPA and percentage ranges
def encode_highschool_graduation_score(value):
    if pd.isna(value) or value == 'Not ortalaması yok':  # Handle NaN and 'Not ortalaması yok'
        return None
    value = value.strip().lower().replace(' ', '')  # Normalize the value by removing spaces and lowercasing
    
    # Highest Range: GPA and percentage ranges
    if value in ['3.00-4.00', '4.00-3.50', '100-85', '75-100']:
        return 4  # Highest range
    
    # Upper-Mid Range: GPA and percentage ranges
    elif value in ['3.50-3.00', '3.50-3', '84-70', '50-75']:
        return 3  # Upper-mid range
    
    # Mid Range: GPA and percentage ranges
    elif value in ['3.00-2.50', '69-55', '50-74', '25-50', '25-49']:
        return 2  # Mid range
    
    # Lowest Range: GPA and percentage ranges
    elif value in ['2.50vealtı', '54-45', '44-0', '0-25', '0-24']:
        return 1  # Lowest range
    
    return None  # For unexpected values

# Apply the encoding function to the 'Lise Mezuniyet Notu' column
cleaned_train_data['Lise Mezuniyet Notu'] = cleaned_train_data['Lise Mezuniyet Notu'].apply(encode_highschool_graduation_score)
```

```python
import numpy as np
import difflib

# List of 81 official Turkish cities
known_cities = [
    'adana', 'adiyaman', 'afyonkarahisar', 'agri', 'aksaray', 'amasya', 'ankara', 'antalya', 'ardahan', 
    'artvin', 'aydin', 'balikesir', 'bartin', 'batman', 'bayburt', 'bilecik', 'bingol', 'bitlis', 
    'bolu', 'burdur', 'bursa', 'canakkale', 'cankiri', 'corum', 'denizli', 'diyarbakir', 'duzce', 
    'edirne', 'elazig', 'erzincan', 'erzurum', 'eskisehir', 'gaziantep', 'giresun', 'gumushane', 
    'hakkari', 'hatay', 'igdir', 'isparta', 'istanbul', 'izmir', 'kahramanmaras', 'karabuk', 
    'karaman', 'kars', 'kastamonu', 'kayseri', 'kilis', 'kirikkale', 'kirklareli', 'kirsehir', 
    'kocaeli', 'konya', 'kutahya', 'malatya', 'manisa', 'mardin', 'mersin', 'mugla', 'mus', 
    'nevsehir', 'nigde', 'ordu', 'osmaniye', 'rize', 'sakarya', 'samsun', 'sanliurfa', 'siirt', 
    'sinop', 'sirnak', 'sivas', 'tekirdag', 'tokat', 'trabzon', 'tunceli', 'usak', 'van', 
    'yalova', 'yozgat', 'zonguldak'
]

# Function to replace Turkish characters with English equivalents
def replace_turkish_chars(city):
    if pd.isna(city):
        return city
    city = city.lower().strip()
    
    # Replace Turkish characters with their English equivalents
    turkish_to_english = str.maketrans({
        'ç': 'c', 'ş': 's', 'ı': 'i', 'ö': 'o', 'ü': 'u', 'ğ': 'g', 'İ': 'i',
        'Ç': 'c', 'Ş': 's', 'Ö': 'o', 'Ü': 'u', 'Ğ': 'g'
    })
    city = city.translate(turkish_to_english)
    
    # Remove district names if present (handle "city/district" format)
    if '/' in city:
        city = city.split('/')[0]
    
    return city

# Function to find closest match from known cities, only if the city is a string
def get_closest_city(city):
    if isinstance(city, str):  # Ensure the value is a string
        # Use difflib to find close matches to known cities
        closest_match = difflib.get_close_matches(city, known_cities, n=1, cutoff=0.6)
        
        # If a close match is found, return it; otherwise, return None (to handle non-matching cases)
        return closest_match[0] if closest_match else None
    return city

# Apply normalization and matching to the 'Ikametgah Sehri' column
cleaned_train_data['Ikametgah Sehri'] = cleaned_train_data['Ikametgah Sehri'].apply(replace_turkish_chars)
cleaned_train_data['Ikametgah Sehri'] = cleaned_train_data['Ikametgah Sehri'].apply(get_closest_city)

# Get the unique values after normalization and matching
unique_cities = cleaned_train_data['Ikametgah Sehri'].unique()

# Create a mapping from city names to unique integers
city_mapping = {city: idx for idx, city in enumerate(unique_cities, start=1) if city is not None}

# Standardize None to NaN for consistency in missing data
cleaned_train_data['Ikametgah Sehri'] = cleaned_train_data['Ikametgah Sehri'].replace([None], np.nan)

# Apply the mapping to the 'Ikametgah Sehri' column (non-matched cities will remain None)
cleaned_train_data['Ikametgah Sehri Encoded'] = cleaned_train_data['Ikametgah Sehri'].map(city_mapping)

# List of unique cities from the 'Ikametgah Sehri' column after processing
unique_cities_after_processing = [city for city in cleaned_train_data['Ikametgah Sehri'].unique() if pd.notna(city)]

# Find cities that are in the processed data but not in the official known cities
extra_cities = [city for city in unique_cities_after_processing if city not in known_cities]

# Find cities in known cities but missing in the processed list
missing_cities = [city for city in known_cities if city not in unique_cities_after_processing]

# Output the extra and missing cities
print(f"Extra cities found: {extra_cities}")
print(f"Missing cities from the official list: {missing_cities}")


# Display the total number of rows to verify no rows are lost
print(f"Total number of rows in the DataFrame: {len(cleaned_train_data)}")

# Output the unique encoded values to verify
print(cleaned_train_data['Ikametgah Sehri Encoded'].unique())
```
